4. [7 pts] Implementation of the circuit for the teleportation of an entangled pair $\frac{1}{\sqrt{2}}(|01\rangle + |10\rangle).$ To do so, we will need a GHZ state.         
Step 1: Prepare a 5 qubit register. The first two qubits are entangled  $\frac{1}{\sqrt{2}}(|01\rangle + |10\rangle).$ The last three are in the GHZ state.     
Step 2: Apply Hadamard on the second qubit, then apply a CX between the second and the third bit.        
Step 3: Apply a gate to the second bit, such that the second qubit recovers the state before step 2.         
Step 4: Apply a CX on the thrid and fourth qubit.         
Step 5: Apply a control phase shift with an angle $\pi$, which is equivalent to $C-\sigma_z$,  on the second and third qubit. Hint: The decomposition of the $ (C-\sigma_z)$ is related to CX with a unitary transformation.   
Step 6: Complete the circuit by applying cx(q[2], q[4]), cx(q[4], q[3]), cx(q[0], q[3]), h(q[4]), cx(q[4], q[0]).     
From your measurement outcome, what is the final wave vector you get? Is the teleportation successful? Why or why not?
